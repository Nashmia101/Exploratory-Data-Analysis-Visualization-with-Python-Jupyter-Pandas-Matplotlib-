**<font color = blue > 2. What is the distribution of remote ratio (0 - No remote work (less than 20%) ; 50 -
Partially remote; 100 - Fully remote (more than 80%) for each of these three
countries? </font>**

```python
USrows = salary[salary['company_location'] == 'US']

zerocount = 0
fiftycount = 0
hundredcount = 0

for ratio in USrows['remote_ratio']:
    if ratio == 0:
        zerocount=zerocount+1
    elif ratio == 50:
        fiftycount=fiftycount+1
    elif ratio == 100:
        hundredcount=hundredcount+1

labels = ['0', '50', '100']
counts = [zerocount, fiftycount, hundredcount]

plt.figure(figsize=(8, 8))
plt.pie(counts, labels=labels)
plt.title('Remote_Ratio Distribution for US Company locations')
plt.show()
print("The number of US comapny locations with 0 remote ratio are:", zerocount)
print("The number of US comapny locations with 50 remote ratio are:", fiftycount)
print("The number of US comapny locations with 100 remote ratio are:", hundredcount)
```

**<font color = purple> Ans:Here I have made use of pie chart to represent the distribution of remote work across the US company locations. The Blue sector in this chart represents 0 (which means less than 20% remote work: No remote work) and The Green sector in this chart represents 100( which means more than 80% remote work: Fully remote) and the Orange sector represents 50( which means partially remote 50 %). Hence from this pie chart we can see that majority (more than half) of the US based companies do not have any remote work while less than half but still a prominant amount of US based companies are fully remote and a small amount of companies are partially remote. Hence from this we can conclude that majority of US based companies are not remote since the blue sector of the circle is larger than green and Orange and US based companies have less partially remote companies since the orange sector is really small.   </font>**

**<font color = blue > 1. Is there any association between Experience Level and the Salary for all jobs?
</font>**

**<font color = orange > 1. Down below I have made 9 grouped bar charts for B3 part 1 by first characterising jobs titles into different categories according to their first letter then I have made a grouped bar chart for every catergory against salary and experience level.
</font>**

```python
A_data = salary[salary['job_title'].str.startswith('A')][['job_title', 'salary_in_usd', 'experience_level']]
A_data
```

```python
pivot_table = A_data.pivot_table(index='experience_level', columns='job_title', values='salary_in_usd', aggfunc='max')

pivot_table = pivot_table[pivot_table.max().sort_values().index]

pivot_table.plot(kind='bar', stacked=False, figsize=(10, 6))

plt.xlabel('Experience Level')
plt.ylabel('Total Salary in USD')
plt.title('Total Salary by Experience Level for Job Titles Starting with "A"')
plt.xticks(rotation=45)
plt.show()
```

**<font color = purple> Ans: For every job title displayed, there seems to be a strong positive correlation between experience level and pay, according to the bar chart. Salary increases generally occur when experience levels rise from entry-level (EN) to executive-level (EX)/ Senior Level (SE). better experience levels are linked to better compensation, as evidenced by the pattern that is consistent across all job titles beginning with "A". The job title "AI Scientist" positions show the biggest increase in salary from entry level to Senior Level</font>**
